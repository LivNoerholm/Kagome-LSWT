```python
T = np.linspace(0,70,141)

def extract_number(filename):
    return int(re.search(r'\d+', filename).group())

folder_path = 'energyarrays'
sorted_files = sorted(os.listdir(folder_path), key=extract_number)

sorted_files
```

```python
E_var_list = []
data_array = []
for filename in sorted_files:
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        data = np.genfromtxt(file_path, delimiter=",")[50:]
        data_array.append(data)
        variance = np.var(data)
        E_var_list.append(variance)

C_v = []
for i in range(len(E_var_list)):
    shc = E_var_list[i]/(T[i]**2)
    C_v.append(shc)

C_v = np.array(C_v)[2:]
T = T[2:]
C_v
```

```python
plt.figure(figsize=(8,5))
plt.plot(T, C_v)
plt.xlabel('Temperature (T)')
plt.ylabel('Specific Heat (C_v)')
plt.title('Specific Heat vs Temperature')
plt.grid()
plt.show()
```

```python
mean_Es = []
for filename in sorted_files:
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        data = np.genfromtxt(file_path, delimiter=",")[55:]
        mean_E = np.mean(data)
        mean_Es.append(mean_E)
plt.figure(figsize=(8,5))
plt.plot(T, mean_Es)
plt.xlabel('Temperature (T)')
plt.ylabel('Mean Energy (E)')
plt.title('Mean Energy vs Temperature')

dE_dt = np.gradient(mean_Es, T)
plt.figure(figsize=(8,5))
plt.plot(T, dE_dt)
plt.xlabel('Temperature (T)')
plt.ylabel('dE/dT')
plt.title('Derivative of Mean Energy vs Temperature')
```
