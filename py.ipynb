```python
import numpy as np
import matplotlib.pyplot as plt
import os
import re
```

```python
data = np.genfromtxt("resdata.txt", delimiter=',')
E = np.genfromtxt("resE.txt")

X = np.linspace(0, 100, len(E))
```

```python
plt.figure(figsize=(10, 7)) 
plt.pcolormesh(X, E, data, vmax=15)
plt.yticks(np.arange(min(E), max(E), 2))  # adjust step size
plt.colorbar()
```

```python
T = np.linspace(0,70, 141)
T[0] = 0.1

def extract_number(filename):
    return int(re.search(r'\d+', filename).group())

folder_path = 'energyarrays'
sorted_files = sorted(os.listdir(folder_path), key=extract_number)

sorted_files
```

```python
E_var_list = []
data_array = []
for filename in sorted_files:
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        data = np.genfromtxt(file_path, delimiter=",")[500:]
        data_array.append(data)
        variance = np.var(data)
        E_var_list.append(variance)

C_v = []
for i in range(len(E_var_list)):
    shc = E_var_list[i]/(T[i]**2)
    C_v.append(shc)
```

```python
plt.figure(figsize=(8,5))
plt.plot(T[2:], C_v[2:])
```
